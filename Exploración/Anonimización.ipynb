```python
# Abrir xlsx usando la ruta definida
import pandas as pd
data = pd.read_excel(r"../Data/Engine_2.xlsx")
data.head()
```

```python
# Anonimizar columnas específicas manteniendo formato categórico
columns_to_anonymize = ['Component ID', 'Asset ID', 'Flota', 'Location', 'Component Profile',
                        'Lubricant', 'Connection Code', 'Observation Code', 'Reviewer']

for column in columns_to_anonymize:
    # Crear categorías únicas
    categories = data[column].astype('category')
    # Crear mapeo con prefijo basado en el nombre de la columna
    prefix = column.replace(' ', '_')
    mapping = {cat: f"{prefix}_{i+1}" for i, cat in enumerate(categories.cat.categories)}
    # Aplicar el mapeo
    data[column] = data[column].map(mapping).astype('category')

data.head()
```

```python
# Renombrar algunas columnas
data = data.rename(columns={'Trakka Rating': 'Rule Based Rating'})

data.head()
```

```python
# Guardar el dataframe anonimizado a un csv
data.to_csv(r"../Data/Data_Anonimizada.csv", index=False)
```
